    3.2. Crie um gráfico que mostra a entropia de um ensaio de Bernoulli (https://pt.wikipedia.org/wiki/Ensaio_de_Bernoulli) para diferentes valores de probabilidade.

```python
def binomial(n, k):
    if 'scipy.special.factorial' not in sys.modules:
        import scipy.special
    if (isinstance(n, (np.integer, int)) == False or isinstance(k, (np.integer, int)) == False):
        raise TypeError("n and k must be integers")
    if k > n:
        raise ValueError("n must be equal or larger then k")
    return scipy.special.factorial(n) / scipy.special.factorial(k) / scipy.special.factorial(n-k)
```

```python
def P(n, k, p):
    return binomial(n, k) * p**k * (1-p)**(n-k)
```

```python
def entropy(n,k):
    p = np.linspace(0,1,100)[1:-1]
    p2 = P(n, k, p)
    ai = np.log2(1/p2)
    e = ai * p
    return p, e
```

```python
np.concatenate((np.arange(1,5), np.arange(1,5)), axis=0)
```

```python
%%opts Curve [height=400 width=400 show_grid=True] (line_color='red')
ns = np.array(list(range(1,4))+list(range(4,10,3))+list(range(10,30,5)))
curve_entropy = {(n,k):hv.Curve(entropy(n,k)).redim(x='p', y='entropy') for n in ns for k in np.arange(0,n+1)}
hv.HoloMap(curve_entropy, kdims=['n', 'k'])
```

```python
%%opts Curve [height=100 width=100]
hv.GridSpace({k:curve_entropy[k] for k in curve_entropy if k[0]<=5}, kdims=['n', 'k'])
```

4. Criar, em qualquer biblioteca, um scatterplot do resultado de um "XOR" (isto é, os pontos (0, 0) e (1, 1) como pertencentes a uma classe, e os pontos (0, 1) e (1, 0) como pertencentes a outra classe).

```python
#generate data
import itertools
x = list(itertools.product((0,1), repeat=2))
df = pd.DataFrame(data=x, columns=['x','y'])
df['xor'] = df.apply(lambda x: x.x ^ x.y, axis=1)
```

```python
%%opts Points [show_grid=True] (size=10)
%%opts NdOverlay [legend_position='right' width=425]

ds = hv.Dataset(df, kdims=['xor'])
points = ds.to(hv.Points, kdims=['x', 'y']).overlay().redim(x={'range':(-0.1,1.1)}, y={'range':(-0.1,1.1)})
points
```

    4.1. Inserir uma curva nesse gráfico que separe os pontos de "classes" distintas.

```python
#This cuvers could be infered by some model (a logistic regression, for instance)
#However, to keep the viz focus, will just input manual curves (x=0.5 and y=0.5).
points * hv.Curve([(0.5,-0.1),(0.5,1.1)]) * hv.Curve([(-0.1,0.5),(1.1,0.5)])
```
